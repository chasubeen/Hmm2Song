```python
# info.csv path 컬럼 붙이기
info['path'] = np.nan

for f in full_input_files:
  num = f[:-4]
  i = int(num)
  info['path'][i] = src + f
```

```python
# path 빈 열 삭제
info_new = info.dropna(subset=['path'])
```

```python
### 데이터형 변경

# id 컬럼을 object 형으로 변경
info_new['id'] = info_new['id'].astype('object')

# id 컬럼을 3자리로 맞춤
info_new['id'] = info_new['id'].apply(lambda x: str(x).zfill(3))
```

## Split_input

```python
# Full_input 내의 모든 파일을 10초 단위로 잘라서 Split_input 폴더에 넣기
## 이때 2초 정도씩 겹치도록 구성 ##
for f in full_input_files:
  f_name = f[:-4]
  song = AudioSegment.from_file(src + f)
  cnt = 0
  for i in range(0, len(song), 8000): # window size는 8초
    tmp_song = song[i:i+10000]
    tmp_song.export('/content/drive/MyDrive/Hmm2Song/data/song/Split_input/' + f_name + "_" + str(cnt) + ".mp3", format='mp3')
    cnt += 1
```
